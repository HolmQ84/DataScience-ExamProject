```python
LyricsGenius.remove_section_headers = True
LyricsGenius.verbose = False

def combineSongAndLyric():
    # Start timer.
    starttimer = time.perf_counter()
    
    trackName = []
    artistName = []
    lyrics = []
    
    
    for i,track in enumerate(sp.playlist_tracks(playlist_URI)["items"]):
    
        #Track name
        track_name = track["track"]["name"]
        #print(track_name)
        
        #Artist name
        artist_name = track["track"]["artists"][0]["name"]
        #print(artist_name)
        
        #Songs Lyrics
        song_lyric = LyricsGenius.search_song(track_name, artist_name).lyrics
        #print(song_lyric)
        
        trackName.append(track_name)
        artistName.append(artist_name)
        lyrics.append(song_lyric)
        
        print(str(i+1) + ' of ' + str(len(sp.playlist_tracks(playlist_URI)["items"])) + ' songs iterated.', end='\r')
    
    tempDf = pd.DataFrame({
    'track':trackName,
    'artist':artistName,
    'lyrics':lyrics    
    })
    
    endtimer = time.perf_counter()
    print(f"Collected lyrics from external links in {endtimer - starttimer:0.4f} seconds")
    
    return tempDf
```
