```python
# Add columns to existing dataset with the audio features.
# Implementing a timer to check the time for adding new columns.
import time
tic = time.perf_counter()

for index, row in reduced_data.iterrows():
    current = sp.audio_features(row['url'])[0]
    reduced_data.at[index,'danceability'] = float(current['danceability'])
    reduced_data.at[index,'energy'] = current['energy']
    reduced_data.at[index,'key'] = current['key']
    reduced_data.at[index,'loudness'] = current['loudness']
    reduced_data.at[index,'mode'] = current['mode']
    reduced_data.at[index,'speechiness'] = current['speechiness']
    reduced_data.at[index,'acousticness'] = current['acousticness']
    reduced_data.at[index,'instrumentalness'] = current['instrumentalness']
    reduced_data.at[index,'liveness'] = current['liveness']
    reduced_data.at[index,'valence'] = current['valence']
    reduced_data.at[index,'tempo'] = current['tempo']
    reduced_data.at[index,'duration_ms'] = current['duration_ms']
    print(str(int(((index+1)/reduced_data.shape[0])*100)) + '% done - ' + str(index+1) + ' of ' + str(reduced_data.shape[0]) + ' songs iterated.', end='\r')
    
toc = time.perf_counter()
print(f"Added new rows from external links in {toc - tic:0.4f} seconds")
```

```python
import re
import tokenize
sentence = "Hejsa hvordan går det med dig, ok det lyder godt"

def word_extraction(sentence):    
    ignore = ['a', "the", "is"]    
    words = re.sub("[^\w]", " ",  sentence).split()    
    cleaned_text = [
        w.lower() for w in words if w not in ignore
    ]    
    return cleaned_text
```

```python
def create_word_dictionary(lyric):
    words=lyric.split(" ")
    word_dictionary = {}
    for word in words:
        if word in word_dictionary:
            word_dictionary[word] = word_dictionary[word]+1
        else:
            word_dictionary[word]=1
    return word_dictionary
```
