```python
import time
tic = time.perf_counter()

for index, row in reduced_data.iterrows():
    current = sp.audio_features(row['url'])[0]
    reduced_data.at[index,'danceability'] = current['danceability']
    reduced_data.at[index,'energy'] = current['energy']
    reduced_data.at[index,'key'] = current['key']
    reduced_data.at[index,'loudness'] = current['loudness']
    reduced_data.at[index,'mode'] = current['mode']
    reduced_data.at[index,'speechiness'] = current['speechiness']
    reduced_data.at[index,'acousticness'] = current['acousticness']
    reduced_data.at[index,'instrumentalness'] = current['instrumentalness']
    reduced_data.at[index,'liveness'] = current['liveness']
    reduced_data.at[index,'valence'] = current['valence']
    reduced_data.at[index,'tempo'] = current['tempo']
    reduced_data.at[index,'duration_ms'] = current['duration_ms']
    print(str(index+1) + ' of ' + str(reduced_data.shape[0]) + ' songs iterated.', end='\r')
toc = time.perf_counter()
print(f"Added new rows from external links in {toc - tic:0.4f} seconds")
```
